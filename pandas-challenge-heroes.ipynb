### Gaming Data Analysis
* Of the active players, the vast majority are male (84%). There also exists, a smaller, but notable proportion of female players (14%).
* Our peak age demographic falls between 20-24 (44.8%) with secondary groups falling between 15-19 (18.6%) and 25-29 (13.4%). 
* The age group that spends the most money is the 20-24 age group.
-----

```python
# Importing pandas Library
import pandas as pd

# File to Load
file_path = "purchase_data.csv"

# Read Purchasing File and store into pandas data frame
raw_data = pd.read_csv(file_path)
raw_data.head()
```

```python
# Calculate total number of players by finding unique screen names
total_players = len(raw_data["SN"].unique())

# Put into new data frame
total_players_df = pd.DataFrame({"Total Players": [total_players]})

# Output data frame
total_players_df
```

```python
# Calculate number of unique items in Item ID column
unique_items = len(raw_data["Item ID"].unique())

# Calculate the total revenue by summing the Price column
total_revenue = sum(raw_data["Price"])

# Calculate average price by diving the sum by the length of the Price column
avg_price = total_revenue/len(raw_data["Price"])

# Calculate number of purchases
num_of_purchases = len(raw_data["Purchase ID"])

# Put into new data frame
purchase_analysis_df = pd.DataFrame({"Number of Unique Items": [unique_items], "Average Price": [avg_price],
                              "Number of Purchases":[num_of_purchases], "Total Revenue":[total_revenue]})

# Format columns with money
purchase_analysis_df["Average Price"] = purchase_analysis_df["Average Price"].map("${:,.2f}".format)
purchase_analysis_df["Total Revenue"] = purchase_analysis_df["Total Revenue"].map("${:,.2f}".format)

# Output data frame
purchase_analysis_df
```

```python
gender_group_df = raw_data.groupby('Gender')

gender_count = gender_group_df["SN"].nunique()
gender_percentage = gender_count/total_players*100
gender_demo_df = pd.DataFrame({"Total Count": gender_count, "Percentage of Players": round(gender_percentage,2)})

gender_demo_df.index.name = None

gender_clndemo_df = gender_demo_df.sort_values(["Total Count"], ascending = False)

gender_clndemo_df
```

```python
gen_purchases = gender_group_df["Purchase ID"].count()
gen_avgpurprice = gender_group_df["Price"].mean()
gen_totalval = gender_group_df["Price"].sum()
gen_eachavg = gen_totalval/gender_count

gen_summary_df = pd.DataFrame({"Purchase Count": gen_purchases,
                                "Average Purchase Price": gen_avgpurprice,
                               "Total Purchase Value": gen_totalval,
                               "Avg Total Purchase per Person": gen_eachavg})

gen_summary_df.index.name = "Gender"

gen_summary_df["Average Purchase Price"] = gen_summary_df["Average Purchase Price"].map("${:,.2f}".format)
gen_summary_df["Total Purchase Value"] = gen_summary_df["Total Purchase Value"].map("${:,.2f}".format)
gen_summary_df["Avg Total Purchase per Person"] = gen_summary_df["Avg Total Purchase per Person"].map("${:,.2f}".format)

gen_summary_df
```

```python
bins = [0, 9.9, 14.9, 19.9, 24.9, 29.9, 34.9, 39.9, 1000]
ages = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

raw_data["Age Groups"] = pd.cut(raw_data["Age"], bins, ages, labels=ages)
age_group_df = raw_data.groupby("Age Groups")

age_count = age_group_df["SN"].nunique()
age_perc = age_count/total_players*100

age_demo_df = pd.DataFrame({"Total Count": age_count,
                           "Percentage of Players": round(age_perc, 2)})

age_demo_df.index.name = None

age_demo_df
```

```python
age_purchases = age_group_df["Purchase ID"].count()
age_avgpurprice = age_group_df["Price"].mean()
age_totalval = age_group_df["Price"].sum()
age_eachavg = age_totalval/age_count

age_summary_df = pd.DataFrame({"Purchase Count": age_purchases,
                                "Average Purchase Price": age_avgpurprice,
                               "Total Purchase Value": age_totalval,
                               "Avg Total Purchase per Person": age_eachavg})

age_summary_df.index.name = None

age_summary_df["Average Purchase Price"] = age_summary_df["Average Purchase Price"].map("${:,.2f}".format)
age_summary_df["Total Purchase Value"] = age_summary_df["Total Purchase Value"].map("${:,.2f}".format)
age_summary_df["Avg Total Purchase per Person"] = age_summary_df["Avg Total Purchase per Person"].map("${:,.2f}".format)

age_summary_df
```

```python
SN_group_df = raw_data.groupby("SN")

SN_purchases = SN_group_df["Purchase ID"].count()
SN_avgpurprice = SN_group_df["Price"].mean()
SN_totalval = SN_group_df["Price"].sum()

SN_summary_df = pd.DataFrame({"Purchase Count": SN_purchases,
                                "Average Purchase Price": SN_avgpurprice,
                               "Total Purchase Value": SN_totalval})

SN_summary_df.index.name = "SN"
SN_cln_summary_df = SN_summary_df.sort_values(["Total Purchase Value"], ascending = False)

SN_cln_summary_df["Average Purchase Price"] = SN_cln_summary_df["Average Purchase Price"].map("${:,.2f}".format)
SN_cln_summary_df["Total Purchase Value"] = SN_cln_summary_df["Total Purchase Value"].map("${:,.2f}".format)

SN_cln_summary_df.head()
```

```python
item_group_df = raw_data.groupby(["Item ID","Item Name"])
item_purchases = item_group_df["Purchase ID"].count()
item_price = item_group_df["Price"].mean()
item_totalval = item_group_df["Price"].sum()

item_popular_df = pd.DataFrame({"Purchase Count": item_purchases,
                               "Item Price": item_price,
                               "Total Purchase Value": item_totalval})

item_popular_df.index.name = ("Item ID")

item_popular_df["Item Price"] = item_popular_df["Item Price"].map("${:,.2f}".format)
item_popular_df["Total Purchase Value"] = item_popular_df["Total Purchase Value"].map("${:,.2f}".format)

itemformatted_df = item_popular_df.sort_values(["Purchase Count"], ascending = False)
itemformatted_df.head()
```

```python
profitable_df = itemformatted_df.head(15)

profitableformatted_df = profitable_df.sort_values("Total Purchase Value", ascending=False)

profitableformatted_df.head(6)

profitableformatted_df.drop([19], inplace=True)

profitableformatted_df.head()
```
